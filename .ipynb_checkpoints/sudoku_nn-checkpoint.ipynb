# First approach to solve sudoku using a simple neural network

```python
import pandas as pd
import numpy as np
import keras
import matplotlib.pyplot as plt
import time
```

## Loading dataset
The next block of code loads the dataset. 

The dataset used can be found in this link: https://www.kaggle.com/bryanpark/sudoku

```python
dt = pd.read_csv("sudoku_dataset/sudoku.csv")
print(dt.shape)
```

## Converte the dataset to dataframe type
This function gets the string stored in the dataset and transforms it into an array of char then reshapes it to a 9x9 matrix. In the end, it returns the matrix as dataframe type. We use this function to display the data in a matrix format instead of a string.

```python
def create_dataframe(string):
    array = list(string)
    reshaped = np.reshape(array, (9,9))
    return pd.DataFrame(reshaped)
```

## Show dataset as dataframe
Here it shows the first data from the dataset. The dataset is divided in "quizzes" and "solutions". The first one shown in the output is the quiz and the second one is the solution. Note that the quiz has zeros in it, which represents the missing number. These positions must be filled with a number from 1 to 9 in order to complete the quiz. 

```python
puz = create_dataframe(dt["quizzes"][0])
sol = create_dataframe(dt["solutions"][0])
print("quiz")
display(puz)
print("solution")
display(sol)
```

## Preprocessing the data
The categorical input data has to be preprocessed before the training in the keras framework. This function gets the string from the dataset transforms it to an array of char than converts each char into an int type then each int becomes an individual array. So it results in an array of array of an individual integer. The function, in the end, uses the function from keras that convert each individual int array to an array of size 10. In this array, the number one has the index of that previous integer and the other numbers are zeros.

```python
from keras.utils import to_categorical

def preprocessing_data(data):
    array = [list(map(int,list(i))) for i in data]
    return to_categorical(array, num_classes=10)
```

## Setting the parameters

```python
epochs = 10
batch = 32
train_size = 60000
test_size = 1000
```

## Dividing the dataset into training and test data
In this cell, the dataset is divided into training set and test set while being preprocessed for the training. Notice that the trainig set has 60000 cases and the test set has 1000. Each case from training and test is a array of size 81 (from 9x9 matrix) with an array of size 10 in each dimension.

```python
x_train = preprocessing_data(dt["quizzes"][:train_size])
y_train = preprocessing_data(dt["solutions"][:train_size])

x_test = preprocessing_data(dt["quizzes"][train_size:train_size+test_size])
y_test = preprocessing_data(dt["solutions"][train_size:train_size+test_size])

print(y_train.shape)
# print(y_train)
```

## Training and Testing
The neural network model is sequential. The first layer flattens the input data, so it transforms the input 81x10 array into 810 array. The next layer is a fully connected layer with 810 neurons. The last layer reshaped the array to the original format, 81x10. The loss function was the categorical_crossentropy. The optimizer was rmsprop. We split the training data in 12.5% for the validation set. The accuracy achieved by this configuration was 80%.

```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Reshape, Dropout, BatchNormalization, Activation

model = Sequential()
model.add(Flatten(input_shape=x_train.shape[1:]))
model.add(Dense(810))
model.add(Reshape((81,10)))
# model.add(BatchNormalization())
model.add(Activation('softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

tbCallBack = keras.callbacks.TensorBoard(log_dir='./logs/{}'.format(int(time.time())), histogram_freq=0, write_graph=True, write_images=True)

history = model.fit(x_train, y_train, validation_split=0.125, epochs=epochs, batch_size=batch, callbacks=[tbCallBack])

score = model.evaluate(x_test, y_test, batch_size=batch)
print(score)
```

## Plotting the history
Here can be seen that the accuracy doesn't improve after 10 epochs and the loss is starting to rise. This means that the model with this configuration won't learn with more epochs and that it overfitted a little.

```python
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

## The Results
Here we use the first case from the test set to display the predicted solution from our trained model. The variable x and y are converted to the original format before the preprocessing step. The variable x is the original quiz and the variable y is the solution, both from the first case from the test set.

Before we put the test case into the model to predict, we first have to reshape the case, adding a new dimension to it. The result is then converted from 81x10 to 81 array and stored into p variable.

In the end, the variables x, y, and p are converted to dataframe and displayed.

```python
x = [np.argmax(i) for i in x_test[0]]
y = [np.argmax(i) for i in y_test[0]]

p = model.predict(x_test[0].reshape(1,81,10))
p = p.reshape((81,10))
p = [np.argmax(i) for i in p]

puz = create_dataframe(x)
sol = create_dataframe(y)
pre = create_dataframe(p)

print("quiz")
display(puz)
print("solution")
display(sol)
print("predicted")
display(pre)
```

## Conclusion
It can be seen that the model got the first line almost right. But unfortunately, it has repeated number in the same column, lines, and square. This means that the model got some number right, but was not able to learn all the rules of the puzzle.
